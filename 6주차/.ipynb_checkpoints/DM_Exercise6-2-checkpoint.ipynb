```python
old_data = data[data['manufact']!="UD"].drop(['model','sales','lnsales','Partition','type'],1)
new_data = data[data['manufact']=="UD"].drop(['model','sales','lnsales','Partition','type'],1).drop(['manufact'],1)
```

```python
new_data['manufact'] = new_pred
```

```python
newCar = list(new_data.loc[152][:-1])
newTruck = list(new_data.loc[153][:-1])
```

#### 제조사 하위 브랜드 간 유사도 거리 구하기(newCar-Saturn / newTruck-Volvo)

```python
def euclidean(v1, v2):
    return sum((p-q)**2 for p, q in zip(v1, v2)) ** .5
```

```python
Saturn = data[data['manufact']=="Saturn"].drop(['sales','lnsales','Partition','type'],1)
Volvo = data[data['manufact']=="Volvo"].drop(['sales','lnsales','Partition','type'],1)
```

```python
Saturn
```

```python
euc_list = []
v1 = newCar
for i in Saturn.values:
    v2 = i[2:]
    eu = euclidean(v1,v2)
    print("model:",i[1])
    print(eu)
    print("======================")
    euc_list.append(eu)
print("최소 유사도 거리:",min(euc_list))
```

```python
euc_list = []
v1 = newTruck
for i in Volvo.values:
    v2 = i[2:]
    eu = euclidean(v1,v2)
    print("model:",i[1])
    print(eu)
    print("======================")
    euc_list.append(eu)
print("최소 유사도 거리:",min(euc_list))
```
