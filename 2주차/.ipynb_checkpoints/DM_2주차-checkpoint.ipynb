## [실습 2-4: KNIME을 활용한 데이터 전처리 - Feature creation and Data transformation]

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

### ■ 가입통신사 공백 제거

```python
data['가입통신사'] = data['가입통신사'].str.replace(" ","")
```

```python
data
```

```python
data['성별'] = data['주민번호'].apply(lambda x: '남' if x[7]=='1' else '여')
```

```python
data
```

```python
data['나이(만)'] = data['주민번호'].apply(lambda x: 2019-int("19"+x[0:2]))
```

```python
data
```

```python
data.describe()
```

```python
data_X = data.drop(["NUM"],1)
```

```python
data_X.corr() ##상관관계 표
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data=data_X.corr(),annot=True,fmt='.2f',cmap='Blues') ##상관관계 시각화해서 보기
```

```python
data_X.boxplot(column='통화료', by='성별')
```
