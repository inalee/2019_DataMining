## DM_3주차 예측 모델링

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("BUYTEST2.csv")
```

```python
data
```

```python
data.isnull().sum() # null 값 확인 
```

```python
compare_str = data['AGE'][9]
for col in data.columns:
    for i in range(len(data[col])):
        if data[col].values[i] == compare_str:
            data[col].values[i] = np.nan
```

```python
data.isnull().sum()
```

```python
# SEX / MARRIED / OWNHOME /ORGSRC 변수의 공백값 최빈값으로 대체


data['MARRIED'] = data['MARRIED'].fillna(data['MARRIED'].mode()[0])
data['SEX'] = data['SEX'].fillna(data['SEX'].mode()[0])
data['OWNHOME'] = data['OWNHOME'].fillna(data['OWNHOME'].mode()[0])
data['ORGSRC'] = data['ORGSRC'].fillna(data['ORGSRC'].mode()[0])
```

```python
data
```

```python
data['AGE'] = data['AGE'].astype('float')
data['INCOME'] = data['INCOME'].astype('float')
data['FICO'] = data['FICO'].astype('float')
```

```python
# AGE / INCOME  / FICO 변수의 공백값은 평균으로 대체
data['AGE'] = data['AGE'].fillna(data[data['AGE'].notna()]['AGE'].mean().round(1))
data['INCOME'] = data['INCOME'].fillna(data[ data['INCOME'].notna()]['INCOME'].mean().round(1))
data['FICO'] = data['FICO'].fillna(data[ data['FICO'].notna()]['FICO'].mean().round(1))
```

```python
data
```

```python
data.isnull().sum() # null 값 확인 
```

```python
data.describe()
```

### ■ 상관관계 파악하기

```python
data2 = data.drop(["ID","C1","C2","C3","C4","C5","C6","C7","PURCHTOT"], 1) ## 필요없는 column들 제거
```

```python
data2.corr()
```

```python
X = data2.iloc[:,1:].values
y = data2.iloc[:,0] ### .ravel()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
x_encoder = ColumnTransformer([('iv_encoder',OneHotEncoder(),[2,6,12]),],remainder='passthrough')
```

```python
X = x_encoder.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from collections import Counter
print(sorted(Counter(y_train).items())) 
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(ratio='auto', kind='regular')

# train데이터를 넣어 복제함
X_resampled, y_resampled = sm.fit_sample(X_train,list(y_train))
```

```python
print(sorted(Counter(y_resampled).items()))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model_1 = DecisionTreeClassifier(criterion='entropy')
model_1.fit(X_resampled,y_resampled)
```

```python
y_pred = model_1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns; 
sns.set()
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test,y_pred) #true , pred 
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.ylabel('true label')
plt.xlabel('predicted label');
```

```python
model_2 = DecisionTreeClassifier(criterion='gini')
model_2.fit(X_resampled,y_resampled)
```

```python
X_pred2 = model_2.predict(X_test)
accuracy_score(y_test,X_pred2)
```

## 3-2: 통신사의 고객 이탈 여부 예측 모델 생성 및 새로운 데이터에 모델 적용

```python
data1 = pd.read_excel("CallsData.xls")
data2 = pd.read_csv("ContractData.csv")
```

```python
data1
```

```python
data2
```

```python
join_data = pd.merge(data1, data2, how='inner',on=['Area Code','Phone'])
```

```python
join_data
```

```python
join_data = join_data.drop(['Phone','Account Length'],1)
```

```python
X = join_data.drop(['Churn'],1)
y = join_data.iloc[:,-4]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
x_encoder = ColumnTransformer([('iv_encoder',OneHotEncoder(),[17]),],remainder='passthrough')
```

```python
X = x_encoder.fit_transform(X)
```

```python
from sklearn.preprocessing import normalize
X = normalize(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
sm = SMOTE(ratio='auto', kind='regular')

X_resampled, y_resampled = sm.fit_sample(X_train,list(y_train))
```

```python
model = DecisionTreeClassifier(criterion='gini')
model.fit(X_resampled,y_resampled)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred) 
```
