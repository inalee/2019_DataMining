```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("BUYTEST.csv")
```

```python
data
```

```python
data.isnull().sum() # null 값 확인 
```

```python
# SEX / MARRIED / OWNHOME /ORGSRC 변수의 공백값 최빈값으로 대체


data['MARRIED'] = data['MARRIED'].fillna(data['MARRIED'].mode()[0])
data['SEX'] = data['SEX'].fillna(data['SEX'].mode()[0])
data['OWNHOME'] = data['OWNHOME'].fillna(data['OWNHOME'].mode()[0])
data['ORGSRC'] = data['ORGSRC'].fillna(data['ORGSRC'].mode()[0])
```

```python
data
```

```python
# AGE / INCOME  / FICO 변수의 공백값은 평균으로 대체

data['AGE'] = data['AGE'].fillna(data[ data['AGE'].notna()]['AGE'].mean().round(1))
data['INCOME'] = data['INCOME'].fillna(data[ data['INCOME'].notna()]['INCOME'].mean().round(1))
data['FICO'] = data['FICO'].fillna(data[ data['FICO'].notna()]['FICO'].mean().round(1))
```

```python
data.isnull().sum() # null 값 확인 
```

```python
data.describe()
```

### ■ 상관관계 파악하기

```python
data2 = data.drop(["ID","C1","C2","C3","C4","C5","C6","C7","PURCHTOT"], 1) ## 필요없는 column들 제거
```

```python
data2.corr()
```

```python
X = data2.iloc[:,1:]
y = data2.iloc[:,0]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
x_encoder = ColumnTransformer([('iv_encoder',OneHotEncoder(),[2,6,12]),],remainder='passthrough')
```

```python
X = x_encoder.fit_transform(X)
```

```python
X[0]
```

```python
len(X[0])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from collections import Counter
print(sorted(Counter(y_train).items())) 
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
print(sorted(Counter(y_resampled).items()))
```

```python
X_resampled
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
model_1 = DecisionTreeClassifier(criterion='entropy')
model_1.fit(X_resampled,y_resampled)
```

```python
X_pred = model_1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns; 
sns.set()
```

```python
accuracy_score(y_test,X_pred)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test,X_pred) #true , pred 
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.ylabel('true label')
plt.xlabel('predicted label');
```
